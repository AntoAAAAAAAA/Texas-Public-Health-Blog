# Health Searches (2005 - 2017)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

Data Cleaning

```python
data = pd.read_csv('Health_Searches.csv')

data = data[data['dma'].str.contains('TX')] #filter for texas rows 

data = data.drop(columns= 'geoCode') #remove geoCode column 

data.head()
```

```python
# data.loc[:, data.columns.str.contains('cancer')]
data_cancer = data.filter(like= 'cancer').copy()
data_cardiovascular = data.filter(like = 'cardiovascular').copy()
data_stroke = data.filter(like= 'stroke').copy()
data_depression = data.filter(like= 'depression').copy()
data_rehab = data.filter(like= 'rehab').copy()
data_vaccine = data.filter(like= 'vaccine').copy()
data_diarrhea = data.filter(like= 'diarrhea').copy()
data_obesity = data.filter(like= 'obesity').copy()
data_diabetes = data.filter(like= 'diabetes').copy()
```

Take mean of cancer for each city, and graph across a bar graph for each city (this can be repeated for other cities)

```python
#re-insert the column that holds location information 
new = data['dma']
data_cancer.loc[:,'Location'] = new 
# move 'Location' to the front index
col = data_cancer.pop('Location')
data_cancer.insert(0, col.name, col)
data_cancer
#Make new dataframe with mean values and Location 
mean = data_cancer.drop(columns= 'Location').mean(axis=1)
summary_df = pd.DataFrame({
    'Location': data_cancer['Location'], 
    'Average': mean
})
#Set 'Location" as the new index and remove rows with NaN values
summary_df = summary_df.set_index('Location')
summary_df = summary_df.dropna()

summary_df.plot(kind='bar', figsize= (10,5))
plt.title('Average Number of Cancer-Related Searches per City in Texas')
plt.legend(['Average Searches'])
plt.xlabel('Cities')
plt.ylabel('Mean Searches')
plt.xticks(rotation=45, ha= 'right')
plt.show()
```

Multi-line graph (line per city) that shows trend of cancer over time 
- repeat for each illness

```python
#Rename columns to just the years
data_cancer.columns = data_cancer.columns.str.slice(0,4)
data_cancer = data_cancer.rename(columns={'Loca':'Location'}) #Revert Location column back to original name

# If 'Location' is already index, reset it first:
if data_cancer.index.name == 'Location':
    data_cancer = data_cancer.reset_index()

# Now safe to set index again:
data_cancer = data_cancer.set_index('Location')
data_cancer_T = data_cancer.T
data_cancer_T.plot(figsize=(15,8))
plt.legend(loc = 'lower right',
           prop= {'size':8})
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Cancer-Related Searches', fontsize = 15)
plt.title('Cancer-Related Searches over time in Texas Cities')
plt.show()

```

```python
%matplotlib notebook

import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import matplotlib.colors as mcolors

# Load shape data
cities = gpd.read_file('tl_2017_48_place.zip')

# Drop irrelevant columns
columns_to_drop = [column for column in data_cancer_T.columns if 'OK' in column]
data_cancer_T = data_cancer_T.drop(columns=columns_to_drop)

# Clean up city names
input_cleaned = [col[:-3].split('-')[0].strip() for col in data_cancer_T.columns]
data_cancer_T.columns = input_cleaned

# Filter matching cities from shapefile
selected_cities = cities[cities['NAME'].isin(input_cleaned)].copy()

# Normalize color scale
vmin = data_cancer_T.min().min()
vmax = data_cancer_T.max().max()
norm = mcolors.Normalize(vmin=vmin, vmax=vmax)

# Set up figure
fig, ax = plt.subplots(figsize=(8, 8))

# Plot function
def update(frame):
    ax.clear()
    cities.plot(ax=ax, facecolor='lightgray', edgecolor='black', linewidth=0.2)

    year_data = data_cancer_T.iloc[frame]
    year_label = data_cancer_T.index[frame]

    merged = selected_cities.copy()
    merged['value'] = merged['NAME'].map(year_data.to_dict())

    merged.plot(
        ax=ax,
        column='value',
        cmap='plasma',
        edgecolor='black',
        linewidth=0.5,
        norm=norm,
        legend=True
    )

    ax.set_title(f"Cancer Data - Year: {year_label}")
    ax.axis('off')

# Create the animation
anim = FuncAnimation(fig, update, frames=len(data_cancer_T), interval=1000, repeat=True)

# This ensures it's shown
plt.show()
```
